```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
import re
```

```python
# Load the dataset
df = pd.read_csv('job_descriptions.csv')
```

```python
df.head()
```

```python
def extract_lower_bound(Experience):
    match = re.search(r'(\d+)', Experience)
    if match:
        return int(match.group(1))
    return None
```

```python
# Apply the extraction function to create a new column for lower bound of experience
df['experience_lower'] = df['Experience'].apply(extract_lower_bound)
```

```python
# Function to extract the lower bound of salary range
def extract_salary_lower_bound(salary_range_str):
    match = re.search(r'(\d+)K', salary_range_str.replace(" − ", " - "))
    if match:
        return int(match.group(1)) * 1000
    return None
```

```python
df['salary_lower'] = df['Salary Range'].apply(extract_salary_lower_bound)
```

```python
# Filter conditions
filtered_df = df[
    (df['Company Size'] < 50000) &
    (df['Job Title'].str.lower() == 'Mechanical Engineer') &
    (df['experience_lower'] <= 5) &
    (df['Country'].str.contains('Asia', case=False)) &
    (df['salary_lower'] > 50000) &
    (df['Work Type'].isin(['Part Time', 'Full Time'])) &
    (df['Preference'].str.lower() == 'Male') &
    (df['Job Portal'].str.lower() == 'Idealist')
]
```

```python
# Check if the filtered DataFrame is empty
if filtered_df.empty:
    print("No data available after applying the filters.")
else:
    print(f"Filtered data contains {len(df)} records.")
    
    # Print a sample of the filtered data
    print(filtered_df.head())

    aggregated_data = filtered_df.groupby('Company')['Company Size'].sum()

    # Plotting
    plt.figure(figsize=(12, 8))
    aggregated_data.plot(kind='bar')
    plt.xlabel('Company Name')
    plt.ylabel('Company Size')
    plt.title('Company Size vs Company Name for Mechanical Engineers')
    plt.xticks(rotation=45)
    plt.show()
```
